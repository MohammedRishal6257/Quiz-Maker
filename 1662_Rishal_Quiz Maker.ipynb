# SIMPLE QUIZ

```python
import tkinter as tk
from tkinter import messagebox

questions = [
    {"question": "Which is the following are valid Tkinter widgets ?", "options": ["ColorPicker", "Label", "WebBrower", "TopLevel"], "answer": "Label"},
        {"question": "What is a primary connection of a router ?", "options": ["Connectivity", "Storage", "Processing", "Output"], "answer": "Connectivity"},
    {"question": "Which language is commonly used for web development ?", "options": ["Python", "Java", "C++", "JavaScript"], "answer": "JavaScript"}
]

question_index = 0
score = 0

def update_question():
    global question_index
    question = questions[question_index]
    question_label.config(text=question["question"])
    
    for i, option in enumerate(question["options"]):
        options_buttons[i].config(text=option, value=option)
    
    option_var.set(None)

def check_answer():
    global question_index, score
    selected_option = option_var.get()
    
    if not selected_option:
        messagebox.showwarning("Warning", "Please select an answer.")
        return
    
    correct_answer = questions[question_index]["answer"]
    selected_answer = [btn.cget("text") for btn in options_buttons if btn.cget("value") == selected_option][0]
    
    if selected_answer == correct_answer:
        score += 1
    
    question_index += 1
    if question_index < len(questions):
        update_question()
    else:
        show_result()

def show_result():
    messagebox.showinfo("Quiz Completed", f"Your score is {score} out of {len(questions)}.")
    root.quit()

root = tk.Tk()
root.title("Quiz Application")

question_label = tk.Label(root, text="", font=("Arial", 14))
question_label.pack(pady=20)

option_var = tk.StringVar()
option_var.set(None)

options_frame = tk.Frame(root)
options_frame.pack(pady=10)

options_buttons = []
for i in range(4):
    btn = tk.Radiobutton(options_frame, text="", variable=option_var, value=i, font=("Arial", 12))
    btn.pack(anchor="w")
    options_buttons.append(btn)

submit_button = tk.Button(root, text="Submit", command=check_answer)
submit_button.pack(pady=20)

update_question()

root.mainloop()
```
